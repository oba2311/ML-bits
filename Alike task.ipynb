```python
import requests
from bs4 import BeautifulSoup
import json
link="https://www.ynet.co.il/home/0,7340,L-184,00.html" 
response = requests.get(link)
soup = BeautifulSoup(response.text,'html.parser') 
headers = soup.findAll('a', {'class': 'smallheader'})
headers

import re
for header in headers:
    clean = re.search("(?<=>).*?(?=<\/a)", str(header))
    print(clean.group(0))
```
